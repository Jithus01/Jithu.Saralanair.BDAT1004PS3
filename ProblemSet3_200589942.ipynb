## Question 1

```python
print("Step 1 - Importing Libraries\n")
import pandas as pd
print(" Step 2 - Fetching URL\n")
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

print(" Step 3 - Assigning to users\n")
users = pd.read_csv(url, delimiter='|')

mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("Step 4 - Discover the mean age per occupation\n")
print("mean_age_per_occupation = \n", mean_age_per_occupation)
```

```python

users['male'] = users['gender'].apply(lambda x: 1 if x == 'M' else 0)
male_ratio = users.groupby('occupation')['male'].mean().sort_values(ascending=False)
print("\nStep 5\n")
print("male_ratio = ",male_ratio)

min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("\nStep 6\n")
print("min_max_age_per_occupation = \n",min_max_age_per_occupation)

mean_age_by_occupation_and_gender = users.groupby(['occupation', 'gender'])['age'].mean()
print("\nStep 7\n")
print("mean_age_by_occupation_and_gender = \n",mean_age_by_occupation_and_gender)

gender_counts = users.groupby('occupation')['gender'].value_counts(normalize=True).unstack()
print("\nStep 8\n")
print("gender_counts = \n",gender_counts)
```

## Question 2

```python
print("Step 1 - Importing Libraries\n")
import pandas as pd

print(" Step 2 - Fetching URL\n")
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv'

print(" Step 3 - Assigning to euro12\n")
euro12 = pd.read_csv(url)

print("\nStep 4 -  Select only the Goal column\n")
goals = euro12['Goals']
print("goals = \n",goals)

print("\nStep 5 - number of team participated in the Euro2012\n")
num_teams = euro12['Team'].nunique()
print("num_teams = \n",num_teams)

print("\nStep 6 - the number of columns in the dataset\n")
num_columns = euro12.shape[1]
print("num_columns = \n",num_columns)

print("\nStep 7 - View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe\n")
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("discipline = \n",discipline)
```

```python

print("\nStep 8 - Sort the teams by Red Cards, then to Yellow Cards\n")
sorted_teams = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)
print("sorted_teams = \n",sorted_teams)

print("\nStep 9 -  the mean Yellow Cards given per Team\n")
mean_yellow_cards = discipline['Yellow Cards'].mean()
print("mean_yellow_cards = \n",mean_yellow_cards)

print("\nStep 10 - teams that scored more than 6 goals\n")
teams_more_than_6_goals = euro12[euro12['Goals'] > 6]
print("teams_more_than_6_goals = \n",teams_more_than_6_goals)

print("\nStep 11 - the teams that start with G\n")
teams_starting_with_G = euro12[euro12['Team'].str.startswith('G')]
print("teams_starting_with_G = \n",teams_starting_with_G)

print("\nStep 12 -  the first 7 columns\n")
first_7_columns = euro12.iloc[:, :7]
print("first_7_columns = \n",first_7_columns)

print("\nStep 13 - all columns except the last 3\n")
all_except_last_3 = euro12.iloc[:, :-3]
print("all_except_last_3 = \n",all_except_last_3)

print("\nStep 14 - only the Shooting Accuracy from England, Italy and Russia\n")
shooting_accuracy_selected_teams = euro12.loc[euro12['Team'].isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']]
print("shooting_accuracy_selected_teams = \n",shooting_accuracy_selected_teams)
```

## Question 4

```python
print("\nStep 1 - Importing Libraries\n")
import pandas as pd
import numpy as np

print("\nStep 2\n - Importing dataset and assigning it to data")

data = pd.read_csv('wind.txt', delim_whitespace=True)

print("\nStep 3 - replace the first 3 columns by a proper datetime index\n")
if data.empty:
    print("\nThe DataFrame is empty. Please check the data loading step.")
else:
    
    print("\nInitial DataFrame:")
    print(data.head())

    data['year'] = data['Yr'] + 1900

    data.rename(columns={'Mo': 'month', 'Dy': 'day'}, inplace=True)

    print("\nColumns after renaming and year adjustment:")
    print(data.columns)

    print("\nYear Range:")
    print(f"Min: {data['year'].min()}, Max: {data['year'].max()}")

    print("\nMonth Range:")
    print(f"Min: {data['month'].min()}, Max: {data['month'].max()}")

    print("\nDay Range:")
    print(f"Min: {data['day'].min()}, Max: {data['day'].max()}")

    valid_years = (data['year'] >= 1900) & (data['year'] <= 2100)
    valid_months = (data['month'] >= 1) & (data['month'] <= 12)
    valid_days = (data['day'] >= 1) & (data['day'] <= 31)

    data = data[valid_years & valid_months & valid_days]

    print("\nFiltered DataFrame:")
    print(data)

    data['Date'] = pd.to_datetime(data[['year', 'month', 'day']], errors='coerce')

    data.dropna(subset=['Date'], inplace=True)

    data.drop(columns=['year', 'month', 'day', 'Yr'], inplace=True)

    data.set_index('Date', inplace=True)

    print("\nFinal DataFrame with Date as index:")
    print(data)

print("\nStep 4 - Create a function to fix error for Year 2061\n")
def fix_year(x):
  if x.year > 1989:
    return x.replace(year=x.year-100)
  else:
    return x

data.index = data.index.map(fix_year)

print("\nStep 5 - Showcasing the corrected index data type\n")
print(data.index.dtype) 

print("\nStep 6 - Computing missing values\n")
missing_values = data.isnull().sum()
print(missing_values)

print("\nStep 7 - Computing non-missing values\n")
non_missing_values = data.notnull().sum().sum()
print(non_missing_values)
```

```python
print("\nStep 8 - Calculating mean windspeed\n")
overall_mean_wind = data.mean().mean()
print(overall_mean_wind)

print("\nStep 9 - calculate the min, max, mean and standard deviations of the windspeeds at each location \n")
loc_stats = data.describe().loc[['min', 'max', 'mean', 'std']]
print(loc_stats)

print("\nStep 10 - calculate the min, max, mean and standard deviations of the windspeeds over each day\n")
day_stats = pd.DataFrame({
    'min': data.min(axis=1),
    'max': data.max(axis=1),
    'mean': data.mean(axis=1),
    'std': data.std(axis=1)
})
print(day_stats.head()) 

print("\nStep 11 - average windspeed in January\n")
january_wind = data[data.index.month == 1].mean()
print(january_wind)

print("\nStep 12 - Downsample the record to a yearly frequency\n")
yearly_wind = data.resample('Y').mean()
print(yearly_wind)

print("\nStep 13 - Downsample the record to a monthly frequency\n")
monthly_wind = data.resample('M').mean()
print(monthly_wind)

print("\nStep 14 - Downsample the record to a weekly frequency\n")
weekly_wind = data.resample('W').mean()
print(weekly_wind)

print("\nStep 15 - calculate the min, max, mean and standard deviations of the windspeeds for each week\n")
weekly_stats = pd.DataFrame({
    'min': weekly_wind.min(axis=1),
    'max': weekly_wind.max(axis=1),
    'mean': weekly_wind.mean(axis=1),
    'std': weekly_wind.std(axis=1)
})

weekly_stats_first_52 = weekly_stats.iloc[:52]
print(weekly_stats_first_52)
```

```python
print("\nStep 1 - Importing libraries\n")
import pandas as pd
import numpy as np
print("\nStep 2 - Import the dataset from this address\n")
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
print("\nStep 3 -Assign it to a variable called chipo\n")
chipo = pd.read_csv(url, sep='\t')
print("\nStep 4 -The first 10 entries\n")
print(chipo.head(10))
print("\nStep 5 -The number of observations in the dataset\n")
print(chipo.shape[0])
print("\nStep 6 -The number of columns in the dataset\n")
print(chipo.shape[1])
print("\nStep 7 -The name of all the columns \n")
print(chipo.columns)
print("\nStep 8 -The dataset indexed\n")
print(chipo.index)
```

```python
print("\nStep 9 -The most-ordered item\n")
most_ordered_item = chipo.groupby('item_name').agg({'quantity': 'sum'}).sort_values('quantity', ascending=False).head(1)
print(most_ordered_item)
print("\nStep 10 -Display items ordered\n")
print(most_ordered_item['quantity'].values[0])
print("\nStep 11 -Display most ordered item in the choice_description column\n")
most_ordered_choice = chipo.groupby('choice_description').agg({'quantity': 'sum'}).sort_values('quantity', ascending=False).head(1)
print(most_ordered_choice)
print("\nStep 12 -Display items ordered in total\n")
total_items_ordered = chipo['quantity'].sum()
print(total_items_ordered)
print("\nStep 13 - \n")
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
print(chipo['item_price'].dtype)
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x))
print(chipo['item_price'].dtype)
print("\nStep 14 -Display revenue for the period\n")
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print(f"Revenue: ${revenue:.2f}")
print("\nStep 15 -Display orders made in the period\n")
total_orders = chipo['order_id'].nunique()
print(total_orders)
print("\nStep 16 -Display average revenue amount\n")
average_revenue_per_order = revenue / total_orders
print(f"Average revenue per order: ${average_revenue_per_order:.2f}")
print("\nStep 17 -Different items sold\n") 
different_items_sold = chipo['item_name'].nunique() 
print("Number of different items sold:", different_items_sold)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'us-marriages-divorces-1867-2014.csv'
data = pd.read_csv(file_path)

print(data.head())

plt.figure(figsize=(10, 6))

plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages per Capita', color='lightblue', linestyle='-', marker='o')

plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces per Capita', color='orange', linestyle='-', marker='o')

plt.xlabel('Year')
plt.ylabel('Per Capita (per 1000 people)')

plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')

plt.legend()

plt.grid(True)

plt.show()
```

## Question 7

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'us-marriages-divorces-1867-2014.csv'
data = pd.read_csv(file_path)

filtered_data = data[(data['Year'] == 1900) | (data['Year'] == 1950) | (data['Year'] == 2000)]

bar_width = 0.35
r1 = range(len(filtered_data))
r2 = [x + bar_width for x in r1]

plt.figure(figsize=(10, 6))

plt.bar(r1, filtered_data['Marriages_per_1000'], color='green', width=bar_width, edgecolor='grey', label='Marriages per Capita')
plt.bar(r2, filtered_data['Divorces_per_1000'], color='indigo', width=bar_width, edgecolor='grey', label='Divorces per Capita')

plt.xlabel('Year', fontweight='bold')
plt.ylabel('Per Capita (per 1000 people)', fontweight='bold')
plt.xticks([r + bar_width/2 for r in range(len(filtered_data))], ['1900', '1950', '2000'])

plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)', fontweight='bold')

plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'roman-emperor-reigns.csv'
data = pd.read_csv(file_path)

assassinated = data[data['Cause_of_Death'] == 'Assassinated'].shape[0]
total = data.shape[0]
other_deaths = total - assassinated

labels = ['Assassinated', 'Other Causes']
sizes = [assassinated, other_deaths]
colors = ['skyblue', 'pink']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})

plt.axis('equal')

plt.title('Fraction of Roman Emperors Assassinated', fontweight='bold')

plt.show()
```
